** Exploratory Data Analysis on Dataset - Terrorism **

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load csv file
df = pd.read_csv("/content/globalterrorismdb_0718dist.csv", encoding="ISO-8859-1")
```

```python
# Group the dataframe by attack type and calculate the mean number of fatalities
attack_fatalities = df.groupby("attacktype1_txt")["nkill"].mean()

# Sort the values in descending order
attack_fatalities = attack_fatalities.sort_values(ascending=False)

# Print the result
print(attack_fatalities)
```

```python
# Group the dataframe by region and count the number of attacks
region_attacks = df.groupby("region_txt")["eventid"].count()

# Sort the values in descending order
region_attacks = region_attacks.sort_values(ascending=False)

# Print the result
print(region_attacks)
```

```python
# Set the figure size
plt.figure(figsize=(10,6))

# Create a bar plot of region_attacks
sns.barplot(x=region_attacks.index, y=region_attacks.values)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.title("Number of Terrorist Attacks by Region")
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(10,6))

# Create a scatter plot of year and number of attacks
sns.scatterplot(x=df["iyear"], y=df["eventid"])

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.title("Number of Terrorist Attacks by Year")

# Show the plot
plt.show()
```
